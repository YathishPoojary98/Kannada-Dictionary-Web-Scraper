```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

base_url = "https://padakanaja.karnataka.gov.in/open_dictionary/%e0%b2%95%e0%b2%a8%e0%b3%8d%e0%b2%a8%e0%b2%a1%20%e0%b2%ae%e0%b2%a4%e0%b3%8d%e0%b2%a4%e0%b3%81%20%e0%b2%b8%e0%b2%82%e0%b2%b8%e0%b3%8d%e0%b2%95%e0%b3%83%e0%b2%a4%e0%b2%bf%20%e0%b2%87%e0%b2%b2%e0%b2%be%e0%b2%96%e0%b3%86/"
page_suffix = "64/200" 
total_pages = 3

with open('scraped_data.txt', 'w', encoding='utf-8') as file:
    for i in range(total_pages):
        url = urljoin(base_url, f"{page_suffix}/{i*200}")
        response = requests.get(url)

        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')

            scraped_data = soup.get_text()

            file.write(scraped_data)

            print(f"Scraping page {i + 1}")
        else:
            print(f"Failed to fetch page {i + 1}")
```

```python
import os
import pandas as pd

# Specify the directory containing the Excel files
folder_path = './Excels/Kannada/'

# List to store dataframes
dfs = []

# Traverse the directory tree
for root, dirs, files in os.walk(folder_path):
    for filename in files:
        if filename.endswith('.xlsx') or filename.endswith('.xls'):  # Check for Excel files
            file_path = os.path.join(root, filename)
            # Read the Excel file
            df = pd.read_excel(file_path)
            dfs.append(df)

# Check if there are any dataframes to concatenate
if dfs:
    # Concatenate all dataframes into one
    combined_df = pd.concat(dfs, ignore_index=True)
    # Example: Display the combined dataframe
    print(combined_df)
    # Optionally, save the combined dataframe to a new Excel file
    #combined_df.to_excel('combined_output.xlsx', index=False)
else:
    print("No valid Excel files found to concatenate.")
```

```python
import os
import pandas as pd
import re

# Folder containing the Excel files
folder_path = "./Excels/English-Kannada/"

# List to store DataFrames from each file
all_data = []

# Iterate through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx") or filename.endswith(".xls"):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_excel(file_path)  # Read each Excel file
        all_data.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_data = pd.concat(all_data, ignore_index=True)

# Initialize a list to store the expanded rows
expanded_rows = []

# Regex to split on both commas and numbered sequences (e.g., '2.', '3.')
split_pattern = re.compile(r'\s*(?:[2-9]|\d{2,})\.\s*|\s*,\s*')

# Loop through each row to split Kannada entries by pattern and create new rows
for _, row in merged_data.iterrows():
    english_word = row['Words']
    kannada_phrases = split_pattern.split(str(row['Meaning']))  # Split using regex
    for phrase in kannada_phrases:
        if phrase.strip():  # Avoid empty entries
            expanded_rows.append((english_word, phrase.strip()))

# Remove duplicates by converting to a set, then back to DataFrame
unique_rows = list(set(expanded_rows))
expanded_data = pd.DataFrame(unique_rows, columns=['English', 'Kannada'])

# Sort the DataFrame by the 'English' column
sorted_data = expanded_data.sort_values(by='English')

# Save the sorted DataFrame into a new Excel file
sorted_data.to_excel("English-Kannada.xlsx", index=False)

print("All Excel files have been merged, expanded, and sorted successfully.")
```
